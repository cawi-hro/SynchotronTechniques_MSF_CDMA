# X-Rays


## What are X-Rays or Rtgenstrahlen

* part of electromagnetic spectrum between ultraviolet (UV) and gamma rays ($\gamma$)
* wavelengths $\lambda$ much shorter than visible light and in atomic scale: $$10^{-11}-10^{-8}\,m\,\,\,\, \text{or} \,\,\,\, 10\,pm-10\,nm$$
* Energies $E$ from: $$100\,eV-100\,keV\,\,\,\, (1\,eV=1.602\cdot 10^{-19}\,J)$$
* frequencies from: $$3\cdot 10^{16}-3\cdot 10^{19}\,Hz\,\,\,\, \text{or} \,\,\,\, 30\,PHz-30\,EHz$$
* high penetration capacity of materials
* can activate substances to fluoresce (emit visible light)

### Short history of X-Rays
* discovered by _Wilhelm Conrad Rntgen_ in 1895
* experimenting with cathode rays  streams of electrons produced in a device called a [_Crookes tube_](https://en.wikipedia.org/wiki/Crookes_tube)
* noticed: unexpected behavior of a screen in distance  starts glowing 
* first picture of his wifes hand

* no patents  further research and development
* 1901 Rntgen was awarded with first _Nobel Prize_ in Physics

## What can we do with X-Rays?

### X-ray Crystallography (XRC)

- [**X-ray Diffraction (XRD)**](https://chem.libretexts.org/Courses/Franklin_and_Marshall_College/Introduction_to_Materials_Characterization__CHM_412_Collaborative_Text/Diffraction_Techniques/X-ray_diffraction_(XRD)_basics_and_application)
    - technique used to determine the atomic arrangement of crystals
    - involves directing X-rays at a crystal and analyzing the resulting diffraction patterns
    - XRD is commonly used in materials science, biology, and chemistry  
      
    [Quelle](https://cdn.serc.carleton.edu/images/NAGTWorkshops/deepearth/activities/silicon_carbide_diffraction_pa.webp)
- **X-ray Fluorescence (XRF)**
    - technique used to determine the chemical composition of materials
    - involves directing X-rays at a sample and analyzing the resulting fluorescence emissions
- **X-ray Absorption Spectroscopy (XAS)**
    - analytical technique used to study the electronic and structural properties of materials by measuring how X-rays are absorbed as a function of energy
    - insights into the local environment of specific atoms within a sample
    - fields: chemistry, materials science, and environmental science to investigate complex systems, catalysts, and metal-organic frameworks.
- **X-ray Diffraction Contrast Tomography (XDCT)**
    - XDCT combines X-ray diffraction and computed tomography to provide 3D imaging of the internal structure of crystalline materials, highlighting variations in crystal orientation and strain
    - This technique enables the visualization of defects and grain boundaries within a sample, making it valuable for materials science and engineering applications
    - XDCT is particularly useful in research focused on understanding the mechanical properties and failure mechanisms of complex materials.

- **Examples**
    - **Structure Determination**:  
        - primarily used to determine the atomic and molecular structure of crystals
        - analyzing how X-rays are diffracted when they interact with the crystal lattice
        - deduce positions of atoms within the crystals
    - **Molecular Biology**:  
        - understanding the structures of biological macromolecules, e.g. proteins and nucleic acids
        - mechanisms of biological processes and drug design
    - **Material Science**:  
        - study the properties of new materials
        - e.g. semiconductors, metals, nanomaterials
        - helping to understand their structure-properties and relationships


### X-ray imaging techniques (XRI)
- diverse range of methods that utilize X-rays to visualize internal structures
- various techniques: radiography, computed tomography (CT), fluoroscopy, and mammography, ...
- multiple fields, including medicine, dentistry, and industrial applications, providing detailed imaging

- **Radiography (XR)**
    - **non-invasive imaging technique** that uses X-rays to create images of the internal structures of the body
    - diagnosing ractures, infections, organs and soft tissues
    - quick and widely available, making it a first-line diagnostic tool in medical settings

- **Computed tomography (CT)** 
    - X-ray source and detectors rotate around subject
    - combines multiple X-ray images taken from different angles to create cross-sectional views
    - detailed information about both bone and soft tissue structures  diagnosing tumors and internal injuries
    - Often used in emergency settings due to its speed and ability to produce comprehensive images quickly.

- **Magnetic Resonance Imaging (MRI)**
    - use of strong magnetic fields and radio waves to generate detailed images of organs and tissues
    - especially effective for imaging soft tissues: brain, spinal cord
    - non-invasive and **does not involve ionizing radiation**  safer for repeated use

- **Dual energy X-ray absorptiometry (DEXA)**
    - specialized X-ray technique used to measure bone mineral density (BMD)
    - diagnose conditions such as osteoporosis by assessing fracture risk.
    - uses two different X-ray energies to differentiate between bone and soft tissue  providing accurate density measurements

- **Fluoroscopy**
    - real-time imaging of internal structures and functions
    - Commonly used in procedures like barium swallows and angiography
    - Helps in diagnosing conditions by visualizing movement and function of organs
    - **Angiography**
        - A medical imaging technique that uses X-rays to visualize blood vessels
        - Typically involves the injection of a contrast dye to enhance visibility
        - Commonly used to diagnose conditions such as blockages, aneurysms, and other vascular issues

- **Mammography**
    - create detailed static images of breast tissue for early detection of breast cancer
    - Typically involves compressing the breast between two plates  clear images while minimizing radiation exposure
    - routine screening tool for women (especially over 40 or with a family history of breast cancer)
